### Gemini Answer Key: Chapter 11 & Lecture Supplement

#### Part 1: Conceptual Problem Set

**1. The Iterative Process**
*   **Why start simple?** Deep learning is highly empirical. You don't know where the actual bottleneck (data, architecture, or hyperparameters) will be until you run the code. Starting with a 'broken' but simple model allows you to establish a baseline and iterate quickly.
*   **High Test Error vs. Training Error:** This indicates **overfitting**. The first step is usually to gather more data or apply regularization (like dropout or weight decay). If the training error is also high, you have **underfitting** and need a larger model or better optimization.

**2. Performance Metrics**
*   **Precision vs. Recall:** In medical diagnosis, **Recall** is usually prioritized (missing a sick patient is worse than a false alarm). In security/spam filters, **Precision** might be prioritized (blocking a legitimate email is worse than letting one spam through).
*   **Single-number evaluation metric:** Metrics like **F1-score** (harmonic mean of precision and recall) allow you to compare two models immediately without debating trade-offs manually every time.

**3. Hyperparameter Prioritization**
*   **Level 1:** Learning rate is the most critical. Others include the hidden layer size and batch size.
*   **Learning Rate Importance:** It controls the effective capacity of the model and whether the optimization converges at all. If it's too high, the model diverges; if too low, it never learns.

**4. Debugging and Visualizing**
*   **Visualizing Mistakes:** Helps identify if the data labels are wrong, if certain classes are being confused due to lack of features, or if the model is focusing on the 'wrong' part of an image.
*   **Overfit a small subset:** If a model cannot achieve zero error on 1-10 examples, there is a fundamental bug in the code (e.g., the loss function or data pipeline) rather than a problem with the learning task.

**5. Error Analysis**
*   By looking at 100 mistakes, you might find that 80% are caused by 'blurry images'. This qualitative insight tells you to focus on 'image sharpening' or 'better camera hardware' (quantitative change) rather than just adding more layers.

**Scenario A (Medical):**
*   **Metric:** Accuracy is misleading due to class imbalance. You should use **Recall** (Sensitivity).
*   **Change:** Use a weighted F-beta score (where beta > 1 to favor recall) or set a minimum recall threshold (e.g., 98%) and then optimize precision.

**Scenario B (Stalling Engine):**
1.  **Step 1: Simplify.** Replace the complex Transformer with a simple Linear Regression or small MLP.
2.  **Step 2: Overfit a tiny subset.** Try to get the loss to zero using only 2-5 product examples.
3.  **Step 3: Check the Learning Rate.** Perform a 'LR Range Test' to ensure the loss moves at all.
